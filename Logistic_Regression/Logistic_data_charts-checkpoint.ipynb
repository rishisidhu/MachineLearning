```python
import matplotlib.pyplot as plt
import numpy as np
import random
from math import log
%matplotlib inline
random.seed(61)
```

```python
m, b = -1, 15
lower, upper = 0, 15
num_points = 15

x1 = [random.randrange(start=1, stop=(num_points-1)) for i in range(num_points)]
x2 = [random.randrange(start=1, stop=(num_points-1)) for i in range(num_points)]

y1 = [random.randrange(start=lower, stop=m*x+b-1) for x in x1]
y2 = [random.randrange(start=m*x+b+1, stop=upper) for x in x2]
```

```python
plt.plot(np.arange(15), m*np.arange(15)+b)
cat1 = plt.scatter(x1, y1, c='blue')
cat2 = plt.scatter(x2, y2, c='red')
point = plt.scatter(8, 7, c='green', marker='x')
plt.xlabel("Tumor Size")
plt.ylabel("Red Color Intensity")
plt.legend((cat1, cat2), ("Non Malignant", "Malignant"), loc='lower left')
plt.savefig('tum.jpg', dpi=72)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
ls = np.linspace(-10, 10, num=100)
y = sigmoid(ls)
```

```python
plt.plot(ls,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.savefig('sigmoid.jpg', dpi=72)
```

```python
def cost_1(x):
    return [-1*log(y) for y in x]
def cost_2(x):
    return [-1*log(1-y) for y in x]
```

```python
x = np.linspace(0.01,0.99, num=50)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x,cost_1(x))
plt.plot(0.1,-log(0.1),'ro')
plt.plot(0.9,-log(0.9),'kx')
plt.xlabel('Hypothesis H')
plt.ylabel('Cost')
plt.xlim(0,1)
plt.ylim(0,5)
plt.title("True Value = 1")
plt.subplot(1,2,2)
plt.plot(x,cost_2(x))
plt.plot(0.1,-log(1-0.1),'ro')
plt.plot(0.9,-log(1-0.9),'kx')
plt.xlabel('Hypothesis H')
plt.ylabel('Cost')
plt.xlim(0,1)
plt.ylim(0,5)
plt.title("True Value = 0")
plt.savefig('y_1.jpg', dpi=72)
```
